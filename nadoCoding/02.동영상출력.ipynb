```python
import cv2
```

```python
# 동영상 파일 출력

cap = cv2.VideoCapture('data/video.mp4')

while cap.isOpened():   # 동영상파일이 올바로 열렸는지?
    ret, frame = cap.read()   # ret : 성공여부 / 현재 프레임을 frame 변수에 저장
    print(ret,frame)
    if not ret:
        print('더이상 가져올 프레임이 없어요')
        break
    cv2.imshow('video', frame)
    
    if cv2.waitKey(25) == ord('q'):
        print('사용자 입력에 의해 종료합니다.')
        break
    
cap.release()   # VideoCapture 객체 해제
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)   #0번째 카메라장치(deviceID)

if cap.isOpened() == False:
# if not cap.isOpened():
    print("카메라 없음")
    exit()  # 프로그램 종료 
    
while True: #무한 반복
    print(cap.isOpened())
    ret, frame = cap.read()
    if not ret:
        break
    
    cv2.imshow('camera', frame)
    if cv2.waitKey(1) == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```
