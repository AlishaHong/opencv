```python
import cv2
```

```python
#   

cap = cv2.VideoCapture('data/video.mp4')

while cap.isOpened():   #   ?
    ret, frame = cap.read()   # ret :  /   frame  
    print(ret,frame)
    if not ret:
        print('   ')
        break
    cv2.imshow('video', frame)
    
    if cv2.waitKey(25) == ord('q'):
        print('   .')
        break
    
cap.release()   # VideoCapture  
cv2.destroyAllWindows()
```
